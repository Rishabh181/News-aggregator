## Generating Headline , Summary ,insights and category from the article

````python
df[["category" , "Headline", "summary" ,"insights" ]] = None
````

````python
import time
````

### Here sleep time is set to 3 because the GOOGLE-1.5-Flash free version has limits to give response for 20 requests per minute.
#### and soo time limit = 60/20 = 3 seconds per requests

````python
for i in range(len(df)):
    if df.loc[i, "article"] != "Access not allowed" :
        text = df.loc[i , "article"]
        
        response_summary = model.generate_content(f"""
            Summarize the following news article in 2-3 sentences:{text}""")
        
        df.loc[i, "summary"] = response_summary.text
        
        time.sleep(3)
````

````python
df["summary"] = df["summary"].str.replace('\n', ' ', regex=False)
````

````python
for i in range(len(df)):
    if df.loc[i, "article"] != "Access not allowed" :
        text = df.loc[i , "article"]
        
        response_category = model.generate_content(f"""
            from the list of above categories {labels} figure out which 3 most relatable type of category does the article {text} may have , give the
            response in a python list form""")
        
        df.loc[i, "category"] = response_category.text
        time.sleep(3)
````

````python
df["category"] = df["category"].str.replace("```python\n", "" , regex = False)
````

````python
df["category"] = df["category"].str.replace("\n```", "" ,regex= False)
````

````python
# df.head(2)
````

````python
for i in range(len(df)):
    if df.loc[i, "article"] != "Access not allowed" :
        text = df.loc[i , "article"]
        response_headline = model.generate_content(f"""
        Give a suitable Provocative and attention-grabbing headline for the article (only one): 
                                          the article is as follows  {text}""")
        df.loc[i, "Headline"] = response_headline.text
        time.sleep(3)
````

````python
for i in range(len(df)):
    if df.loc[i, "article"] != "Access not allowed" :
        text = df.loc[i , "article"]
        
        response_insights = model.generate_content(f"""
            find the key insights and takeaways from the article :
                                                {text}""")
        df.loc[i, "insights"] = response_insights.text
        time.sleep(3)
````
