# For Categorization of News-article type , Using zero-shot-classification model

```python
import torch
```

```python

model = "facebook/bart-large-mnli" 
```

```python
from transformers import pipeline

# Load zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model= model )
```

```python
# News-categories
labels = [
    "World", "National", "Politics", "Business", "Finance", "Economy", "Technology", 
    "Science", "Health", "Sports", "Entertainment", "Celebrities", "Lifestyle", 
    "Travel", "Food & Drink", "Education", "Environment", "Climate", "Religion", 
    "Spirituality", "Law", "Crime", "Justice", "Opinion", "Editorial", "Weather", 
    "Real Estate", "Fashion", "Style", "Automobiles", "Transportation", 
    "Art & Culture", "Jobs", "Careers", "Gaming", "eSports", "Parenting", 
    "Family", "Obituaries", "Comics", "Horoscopes"
]
```

### Taking most 3 relatable categories with respect to the article

```python
df[["category-1" ,"category-2" ,"category-3"]] = None
```

```python
def categorizer(text):
    result = classifier(text , candidate_labels=labels, multi_label=True )
    return result["labels"][0], result["labels"][1] , result["labels"][2]
```

```python
for i in range(len(df)):
    article = df.loc[i, "article"]
    if (article != "Access not allowed" and article != "" and article != " " and pd.notna(article) ):
       cat1 , cat2 , cat3 = categorizer(df.loc[i , "article"])
       df.loc[i, "category-1"] = cat1
       df.loc[i , "category-2"] = cat2
       df.loc[i , "category-3"] = cat3
```

# Using t5-small (HuggingFace model) for Text summarization and insights extraction.

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, AutoTokenizer

model_name = 't5-small'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
def summarize(text):
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
def insights_generation(text):
    inputs = tokenizer.encode(f"Extract minimum 4 key takeaways from the article in bullet points: {text}", return_tensors="pt", max_length=2000, truncation=True)
    summary_ids = model.generate(inputs, max_length=150, min_length=40, length_penalty=2.0, num_beams=4, no_repeat_ngram_size=2, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
df["summary"] = None
df["insights"] = None
```

```python
for i in range(len(df)):
    if df.loc[i, "article"] != "Access not allowed":
        sum = summarize(df.loc[i , 'article'])
        insights = insights_generation(df.loc[i , 'article'])
        df.loc[i, "summary"] = sum
        df.loc[i, "insights"] = insights
```

## For Suitable Headline Generation using "JulesBelveze/t5-small-headline-generator" pre-trained model

```python
import re
from transformers import T5Tokenizer, T5ForConditionalGeneration, AutoTokenizer

model_name_headline = "JulesBelveze/t5-small-headline-generator"


tokenizer = AutoTokenizer.from_pretrained(model_name_headline)
model = T5ForConditionalGeneration.from_pretrained(model_name_headline)
```

```python
df["Headline"] = None
```

```python
def headline_generator(text):
    input_ids = tokenizer( f"Give a suitable Provocative and attention-grabbing headline for the article (only one) : {text}", return_tensors="pt", padding="max_length", truncation=True, max_length=100)["input_ids"]

    output_ids = model.generate(input_ids=input_ids, max_length=84, no_repeat_ngram_size=2, num_beams=4)[0]

    Headline = tokenizer.decode(output_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False )

    return Headline
```

```python
for i in range(len(df)):
    if df.loc[i, "article"] != "Access not allowed":
        headline = headline_generator(df.loc[i , 'summary'])
        df.loc[i , "Headline"] = headline
```
